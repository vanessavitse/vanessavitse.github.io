**Comparaisons asymptotiques m√©thodes de calcul de Fibonacci**

```python
A=Zmod(101)
def Fib1(n):
    if n==0 or n==1:
        return A(1)
    else:
        return Fib1(n-1)+Fib1(n-2)

def Fib2(n):
    F0,F1=A(1),A(1)
    for i in range(2,n+1):
        F0,F1=F1,F1+F0
    return F1

def Fib3(n):
    M=(matrix([[A(1),A(1)],[A(1),A(0)]]))^n
    return M[1,0]+M[1,1]

Fib3(10)==Fib2(10)==Fib1(10)
```

```python
from time import time
L1=[]
L2=[]
L3=[]
for i in range(30):
    t0=time()
    Fib1(i)
    t1=time()
    L1.append((i,t1-t0))
for i in range(200):
    t0=time()
    Fib2(i)
    t1=time()
    L2.append((i,t1-t0))
    t0=time()
    Fib3(i)
    t1=time()
    L3.append((i,t1-t0))
list_plot(L1,color='red')+list_plot(L2,color='blue')+list_plot(L3,color='green')
```
